## 使用 skimage 计算 SSIM

```python
import cv2
import matplotlib.pyplot as plt
from skimage import metrics
import torch as pt
```

```python
img = cv2.imread(r"C:\Users\a9214\Downloads\duck.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = pt.tensor(img) / 255
plt.imshow(img)
```

### 进行变换

```python
imgs = []
```

```python
imgs[0] = (img + (pt.rand(img.shape)*0.1-0.05).clamp(0, 1)) # 噪点
imgs[1] = (img.flip((0, 1))) # 翻转
```

```python
plt.imshow(imgs[1])
```

```python
metrics.structural_similarity(img.numpy(), imgs[1].numpy(), multichannel=True)
```
